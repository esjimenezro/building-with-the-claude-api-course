## 1. Response Streaming

The API responses may last a bit longer than what a user may be willing to wait. One solution is to put a spinner on the screen.

Another solution is to enable streaming of responses to perceive as if the API were "writing".

```python
# One option
messages = []

add_user_message(
    messages,
    "Write a 1 sentence description of a fake database"
)

stream = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=messages,
    stream=True
)

for event in stream:
    print(event)
```

```python
# Other option
messages = []

add_user_message(
    messages,
    "Write a 1 sentence description of a fake database"
)

with client.messages.stream(
    model=model,
    max_tokens=1000,
    messages=messages
) as stream:
    for text in stream.text_stream:
        # print(text, end="")
        pass

stream.get_final_message()
```
