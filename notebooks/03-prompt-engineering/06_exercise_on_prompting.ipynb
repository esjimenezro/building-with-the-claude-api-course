````python
# Define and run the prompt you want to evaluate, returning the raw model output
# This function is executed once for each test case
def run_prompt(prompt_inputs: dict) -> str:
    prompt = f"""
Extract the topics mentioned in the following passage of text from a scholarly article into a JSON array of strings.

<scholarly_article>
{prompt_inputs["content"]}
</scholarly_article>

Guidelines:
1. Include only a JSON array of strings, containing each topic mentioned in the article.
2. The strings should contain only a topic without any extra commentary.
3. Response should contain the JSON array and nothing else.

Here's an example with a sample input and an ideal output:
<sample_input>
This randomized controlled trial examined the effects of cognitive behavioral therapy on anxiety symptoms in college students. Participants (N=120) were randomly assigned to either the treatment group receiving 12 weeks of CBT or a waitlist control condition. The Beck Anxiety Inventory served as the primary outcome measure, administered at baseline, post-treatment, and 3-month follow-up. Results indicated a significant between-groups effect size (Cohen's d = 0.82, p < .001), with the treatment group showing clinically meaningful reductions in anxiety scores. A repeated measures ANOVA revealed significant time × condition interactions, suggesting sustained therapeutic benefits. The study employed intention-to-treat analysis to address participant attrition, which occurred at a rate of 15% across conditions.
</sample_input>
<ideal_output>
[
  "randomized controlled trial",
  "cognitive behavioral therapy",
  "anxiety symptoms",
  "college students",
  "treatment group",
  "waitlist control condition",
  "Beck Anxiety Inventory",
  "baseline measurement",
  "post-treatment assessment",
  "follow-up assessment",
  "effect size",
  "Cohen's d",
  "repeated measures ANOVA",
  "time × condition interactions",
  "therapeutic benefits",
  "intention-to-treat analysis",
  "participant attrition"
]
</ideal_output>

The solution is an ideal output as it meets all mandatory requirements by providing a clean JSON array of strings with topic extraction only. It successfully extracts psychology-specific topics, statistical terms, and research methodology concepts from the scholarly article. All topics are relevant and properly formatted as strings without commentary.
"""
    
    messages = []
    add_user_message(messages, prompt)
    add_assistant_message(messages, "```json")
    return chat(
        messages=messages,
        client=client,
        model=model,
        stop_sequences=["```"]
    )
````

````python
results = evaluator.run_evaluation(
    run_prompt_function=run_prompt,
    dataset_file="dataset_exercise.json",
    json_output_file="output_exercise.json",
    html_output_file="output_exercise.html",
    extra_criteria="""
- Contains a JSON array of strings, containing each topic mentioned in the article.
- The strings should contain only a topic without any extra commentary.
- Response should contain the JSON array and nothing else.
"""
)
````
